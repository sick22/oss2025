```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'DejaVu Sans'


data = pd.read_csv('/content/Student_Mental_Stress_and_Coping_Mechanisms.csv')


```

```python
# ---------------------------
# (1) 스트레스 수준별 학생 수
# ---------------------------
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Mental Stress Level', order=data['Mental Stress Level'].value_counts().index)
plt.title('Number of Students by Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Number of Students')
plt.show()





```

```python
# ---------------------------
# (2) 주요 스트레스 원인 분포
# ---------------------------
plt.figure(figsize=(8, 8))
# 'Financial Stress', 'Peer Pressure', 'Relationship Stress' 등을 합치는 것도 가능
stress_causes = ['Financial Stress', 'Peer Pressure', 'Relationship Stress']
cause_sum = data[stress_causes].sum()
cause_sum.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Major Stress Causes')
plt.ylabel('')
plt.show()

```

```python
# ---------------------------
# (3) 스트레스 수준에 따른 평균 수면 시간
# ---------------------------
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Mental Stress Level', y='Sleep Duration (Hours per night)', order=data['Mental Stress Level'].value_counts().index)
plt.title('Sleep Duration by Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Sleep Duration (Hours)')
plt.show()


```

```python
# ---------------------------
# (4) 대처 방법별 선호도
# ---------------------------
plt.figure(figsize=(10, 6))
sns.countplot(data=data, y='Stress Coping Mechanisms', order=data['Stress Coping Mechanisms'].value_counts().index)
plt.title('Preferred Coping Mechanisms')
plt.xlabel('Number of Students')
plt.ylabel('Coping Mechanism')
plt.show()
```
